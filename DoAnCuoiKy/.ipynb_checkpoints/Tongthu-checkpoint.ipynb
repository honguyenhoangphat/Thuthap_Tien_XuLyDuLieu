```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('total_revenue.csv')
print(df)
```

```python
# lọc bỏ 2008 - 2018
df = df[~df['Năm'].between(2008, 2018)]
print(df)
print(df.info())
```

```python
df['Tốc độ tăng trưởng (%)'] = df['Tốc độ tăng trưởng (%)'].str.replace(',', '.').astype(float)
print(df.info())
```

```python
# Vẽ biểu đồ xu hướng tổng thu từ du lịch qua các năm
plt.figure(figsize=(12, 7))
plt.plot(df['Năm'], df['Tổng thu từ khách du lịch (nghìn tỷ đồng)'], marker='o', color='b')
plt.title('Xu hướng tổng thu từ khách du lịch qua các năm')
plt.xlabel('Năm')
plt.ylabel('Tổng thu (nghìn tỷ đồng)')
plt.grid(True)
plt.show()

# Vẽ biểu đồ tốc độ tăng trưởng qua cácnămm
plt.figure(figsize=(12, 7))
plt.bar(df['Năm'], df['Tốc độ tăng trưởng (%)'], color='orange')
plt.title('Tốc độ tăng trưởng du lịch qua các năm')
plt.xlabel('Năm')
plt.ylabel('Tốc độ tăng trưởng (%)')
plt.grid(axis='y')
plt.ylim(min(df['Tốc độ tăng trưởng (%)']) - 10, max(df['Tốc độ tăng trưởng (%)']) + 10)  # Điều chỉnh trục y
plt.show()
```
