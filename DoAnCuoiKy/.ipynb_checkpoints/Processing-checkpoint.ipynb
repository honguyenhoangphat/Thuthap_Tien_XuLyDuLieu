```python
import pandas as pd
```

```python
foreign_tourist = pd.read_csv('foreign_tourist.csv')
```

```python
foreign_tourist = foreign_tourist.drop(columns=['Unnamed: 8', 'Unnamed: 9'])
```

```python
foreign_tourist.head()
```

```python
foreign_tourist.dtypes
```

```python
numeric_cols = ['Ước tính mỗi tháng', 'Tổng lượt khách cả năm']
for col in numeric_cols:
    foreign_tourist[col] = foreign_tourist[col].str.replace('.', '').astype(float)

# Thêm cột phân loại
foreign_tourist['Loại du khách'] = 'Quốc tế'
```

```python
foreign_tourist.head()
```

```python
in_tourist = pd.read_excel('in_tourist.xlsx',dtype=str)
in_tourist.head()
```

```python
in_tourist_melted = pd.melt(in_tourist, 
                            id_vars=['Năm', 'Chỉ tiêu'], 
                            value_vars=[f'Tháng {i}' for i in range(1, 13)], 
                            var_name='Tháng', 
                            value_name='Số lượng')
in_tourist_melted['Tháng'] = in_tourist_melted['Tháng'].str.replace('Tháng ', 't')
in_tourist_melted['Loại du khách'] = 'Nội địa'
in_tourist_melted.head()
```

```python
tourist_data = pd.concat([foreign_tourist,in_tourist_melted], ignore_index=True)
tourist_data['Số lượng'] = tourist_data['Ước tính mỗi tháng'].combine_first(tourist_data['Số lượng'])
tourist_data = tourist_data.drop(columns=['Ước tính mỗi tháng'])
tourist_data
```

```python
gdp_data = pd.read_csv('gdp_20192024.csv')
gdp_data
```

```python
gdp_data = gdp_data.rename(columns={'Year': 'Năm'})
gdp_data['Tháng'] = 't12'  # GDP thường là dữ liệu theo năm, đại diện tháng 1
gdp_data
```

```python
import matplotlib.pyplot as plt

# Vẽ biểu đồ GDP
plt.figure(figsize=(10, 6))
plt.plot(gdp_data['Năm'], gdp_data['GDP (Billion USD)'], marker='o', linestyle='-', color='g', label='GDP (Billion USD)')
plt.title('Xu hướng GDP giai đoạn 2019-2024')
plt.xlabel('Năm')
plt.ylabel('GDP (Billion USD)')
plt.grid(True)
plt.xticks(gdp_data['Năm'])
plt.legend();
```

```python
inflation_data = pd.read_csv('inflation_rate.csv')
unemployment_data = pd.read_csv('unemployment.csv')
inflation_data.head(), unemployment_data.head()
```

```python
inflation_data[['Tháng', 'Năm']] = inflation_data['Date'].str.split(' ', expand=True)
inflation_data['Năm'] = inflation_data['Năm'].astype(int)

# Loại bỏ ký hiệu % và chuyển sang dạng float
inflation_data['Inflation Rate'] = inflation_data['Inflation Rate'].str.replace('%', '').astype(float)
inflation_data = inflation_data.rename(columns={'Inflation Rate': 'Inflation Rate (%)'})

# Lọc dữ liệu từ 2019 đến 2024
inflation_filtered = inflation_data[(inflation_data['Năm'] >= 2019) & (inflation_data['Năm'] <= 2024)]

# Đổi tên tháng từ tiếng Anh sang t1, t2,...
month_mapping = {
    'Jan': 't1', 'Feb': 't2', 'Mar': 't3', 'Apr': 't4', 'May': 't5', 'Jun': 't6',
    'Jul': 't7', 'Aug': 't8', 'Sep': 't9', 'Oct': 't10', 'Nov': 't11', 'Dec': 't12'
}
# Áp dụng mapping
inflation_filtered['Tháng'] = inflation_filtered['Tháng'].map(month_mapping)

inflation_filtered = inflation_filtered[['Năm', 'Tháng', 'Inflation Rate (%)']]

inflation_filtered
```

```python
# Vẽ biểu đồ lạm phát
plt.figure(figsize=(10, 6))
for year in inflation_filtered['Năm'].unique():
    data = inflation_filtered[inflation_filtered['Năm'] == year]
    plt.plot(data['Tháng'], data['Inflation Rate (%)'], marker='o', label=f'Lạm phát {year}')
plt.title('Xu hướng tỷ lệ lạm phát giai đoạn 2019-2024')
plt.xlabel('Tháng')
plt.ylabel('Tỷ lệ lạm phát (%)')
plt.grid(True)
plt.legend();
```

```python
inflation_filtered['Inflation Rate (%)'] = inflation_filtered['Inflation Rate (%)'].astype(float)

# Vẽ lại biểu đồ Area (Vùng)
plt.figure(figsize=(12, 6))
for year in inflation_filtered['Năm'].unique():
    data = inflation_filtered[inflation_filtered['Năm'] == year]
    plt.fill_between(data['Tháng'], data['Inflation Rate (%)'], alpha=0.5, label=f'Lạm phát {year}')

plt.title('Xu hướng tỷ lệ lạm phát giai đoạn 2019-2024 (Area Chart)')
plt.xlabel('Tháng')
plt.ylabel('Tỷ lệ lạm phát (%)')
plt.xticks(range(1, 13), [f't{i}' for i in range(1, 13)])
plt.legend()
plt.grid(True)
```

```python
# Lọc dữ liệu thất nghiệp của cả nước
unemployment_filtered = unemployment_data[unemployment_data['Tỉnh/Thành phố'] == 'CẢ NƯỚC']
unemployment_filtered = unemployment_filtered[['Tỉnh/Thành phố', '2019', '2020', '2021', '2022', 'Sơ bộ 2023']]

# Chuyển đổi dữ liệu từ dạng cột sang dòng
unemployment_melted = pd.melt(unemployment_filtered, 
                              id_vars=['Tỉnh/Thành phố'], 
                              value_vars=['2019', '2020', '2021', '2022', 'Sơ bộ 2023'],
                              var_name='Năm', 
                              value_name='Tỷ lệ thất nghiệp')
# Đổi tên
unemployment_melted = unemployment_melted.rename(columns={'Tỷ lệ thất nghiệp': 'Tỷ lệ thất nghiệp (%)'})
#Thêm cột
new_row = pd.DataFrame({'Năm': [2024], 'Tỷ lệ thất nghiệp (%)': ['2,26'], 'Tháng': ['t12']})
unemployment_melted = pd.concat([unemployment_melted, new_row], ignore_index=True)
# Chuyển đổi dữ liệu thất nghiệp về số
unemployment_melted['Tỷ lệ thất nghiệp (%)'] = unemployment_melted['Tỷ lệ thất nghiệp (%)'].str.replace(',', '.').astype(float)

# Đổi tên năm 2023 và thêm 2024
unemployment_melted['Năm'] = unemployment_melted['Năm'].replace({'Sơ bộ 2023': '2023'}).astype(int)
unemployment_melted = unemployment_melted[['Năm', 'Tỷ lệ thất nghiệp (%)']]

unemployment_melted['Tháng'] = 't12'  # Đưa dữ liệu thất nghiệp vào cuối năm
```

```python
unemployment_melted
```

```python
# Vẽ biểu đồ xu hướng tỷ lệ thất nghiệp
plt.figure(figsize=(10, 6))
plt.plot(unemployment_melted['Năm'], unemployment_melted['Tỷ lệ thất nghiệp (%)'],color='Purple', marker='o', linestyle='-', label='Tỷ lệ thất nghiệp (%)')
plt.title('Xu hướng tỷ lệ thất nghiệp giai đoạn 2019-2024')
plt.xlabel('Năm')
plt.ylabel('Tỷ lệ thất nghiệp (%)')
plt.grid(True)
plt.xticks(unemployment_melted['Năm']);
```
