```python
import numpy as np
import pandas as pd
from scipy.stats import stats
from scipy.stats import norm, t
import scipy
from scipy import stats
from scipy.stats import ttest_ind  
```

```python
df = pd.read_csv('Inf.Sal.csv')
df
```

```python
clean = df.drop(columns=['Unnamed: 0'])
clean
```

```python
hcm = df['HCM'].dropna()
hn = df ['HaNoi'].dropna()
```

```python
mean_hcm = np.mean(hcm)
mean_hn = np.mean(hn)
diff_mean = mean_hcm - mean_hn

var_hcm = np.var(hcm, ddof=1)
var_hn = np.var(hn, ddof=1)

se_diff = np.sqrt(var_hcm / len(hcm) + var_hn / len(hn))

ci_lower, ci_upper = norm.interval(0.95, loc=diff_mean, scale=se_diff)
diff_mean, (ci_lower, ci_upper)
```

```python
p_hcm = sum(hcm > 11.5) / len(hcm)
p_hn = sum(hn > 11.5) / len(hn)

t_stat, p_value = stats.ttest_ind(hcm, hn, alternative='greater')
print(p_hcm, p_hn, p_value)
ketqua = 'Bác bỏ giả thuyết' if p_value < 0.025 else 'Chưa đủ cơ sở để bác bỏ'
print(ketqua)
```

```python
def prop_test_geq(x, y, alpha):
    p_x = sum(x > 11.5) / len(x)
    p_y = sum(y > 11.5) / len(y)
    
    z = (p_x - p_y) / ((p_x * (1 - p_x) / len(x)) + (p_y * (1 - p_y) / len(y)))**0.5
    
    p_value = 1 - stats.norm.cdf(z)
    
    print(f"P_value: {p_value:.4f}")
    
    if p_value < alpha:
        print('Kết luận: Bác bỏ H0')
    else:
        print("Kết luận: Chưa đủ cơ sở để bác bỏ")
prop_test_geq(hcm, hn, 0.025)
```

```python
def prop_test_leq(x, y, alpha):
    p_x = sum(x > 11.5) / len(x)
    p_y = sum(y > 11.5) / len(y)
    
    z = (p_x - p_y) / ((p_x * (1 - p_x) / len(x)) + (p_y * (1 - p_y) / len(y)))**0.5
    
    p_value = stats.norm.cdf(z)
    
    print(f"P_value: {p_value:.4f}")
    if p_value < alpha:
        print('Kết luận: Bác bỏ H0')
    else:
        print("Kết luận: Chưa đủ cơ sở để bác bỏ")
prop_test_leq(hcm, hn, 0.025)
```
